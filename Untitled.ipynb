```python
import pandas as pd 
import geopandas as gpd 
import contextily as ctx 
import matplotlib.pyplot as plt 
```

```python
# ctk means columns to keep 
ctk = [ 
    "cat1",
    "Name",
    "addrln1",
    "latitude",
    "longitude",
    "POINT_X",
    "POINT_Y",
    "geometry"
    
]
```

```python
hcs = hcs[ctk]
```

```python
hcs.head()
```

```python
hcs_web_mercator = hcs.to_crs(epsg = 3857)
```

```python
fig, ax = plt.subplots(figsize = (15,15))
hcs_web_mercator.plot(ax=ax, alpha = 0.8)

ax.axis("off")
ax.set_title("title")
ctx.add_basemap(ax)
```
